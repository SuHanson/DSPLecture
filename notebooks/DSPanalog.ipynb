## 基本概念
類比訊號(Analog Signals)定義為「隨時間改變的連續訊號」，亦稱連續時間訊號(Continuous-Time Signals)  
以函數形式表示類比訊號 $x=f(t),\; t\in R$

## 弦波
- 頻率 $f$ (cycles/second)
- 角頻率 $w=2\pi f$ (radians/second)
- 週期 $T=\frac{1}{f}$ (second)

AI提示：幫我產生並繪製一組振幅為1，頻率為2，相位移分別是0，45及90度的弦波

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
A = 1         # Amplitude
f = 2         # Frequency

# Phase shifts in degrees
phase_shifts_deg = [0, 45, 90]

# Create a time array
t = np.linspace(0, 2, 500)  # From 0 to 2 seconds, with 500 points

# Plot the sine waves
plt.figure(figsize=(10, 6))

for phi_deg in phase_shifts_deg:
    phi_rad = np.deg2rad(phi_deg)  # Convert phase shift from degrees to radians
    y = A * np.sin(2 * np.pi * f * t + phi_rad)
    plt.plot(t, y, label=f'Phase Shift = {phi_deg}°')

plt.title('Sine Waves with Different Phase Shifts')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5) # Add a horizontal line at y=0
plt.axvline(0, color='black', linewidth=0.5) # Add a vertical line at x=0
plt.show()
```

### 複數直角座標與極座標轉換

複數可以用兩種主要形式表示：
1.  **直角座標形式 (Rectangular Form)**：`a + bj`，其中 `a` 是實部，`b` 是虛部。
2.  **極座標形式 (Polar Form)**：`r * (cos(θ) + j * sin(θ))`，其中 `r` 是模數 (magnitue)，`θ` 是輻角 (phase angle)。

`cmath` 模組提供了方便的函數來進行這兩種形式之間的轉換。


```python
# 直角座標轉極座標 (rectangular to polar)
# cmath.polar(z) 返回 (r, phi)，r 為模數，phi 為輻角 (弧度)
def rect_to_polar(z: complex):
    r, phi = cmath.polar(z)
    print(f"直角座標 {z} 轉換為極座標：模數 r = {r:.4f}, 輻角 phi = {math.degrees(phi):.4f}° ({phi:.4f} 弧度)")
    return r, phi

# 極座標轉直角座標 (polar to rectangular)
# cmath.rect(r, phi) 返回直角座標複數
def polar_to_rect(r: float, phi: float):
    z = cmath.rect(r, phi)
    print(f"極座標 (r={r:.4f}, phi={math.degrees(phi):.4f}°) 轉換為直角座標：{z:.4f}")
    return z

# 範例用法
z1 = 3 + 4j
print("\n--- 直角座標轉極座標範例 ---")
r1, phi1 = rect_to_polar(z1)

z2 = -1 + 1j
r2, phi2 = rect_to_polar(z2)

print("\n--- 極座標轉直角座標範例 ---")
p_r = 5
p_phi_deg = 53.13
p_phi_rad = math.radians(p_phi_deg)
polar_to_rect(p_r, p_phi_rad)

p_r2 = math.sqrt(2)
p_phi_deg2 = 135
p_phi_rad2 = math.radians(p_phi_deg2)
polar_to_rect(p_r2, p_phi_rad2)
```

### 複數的四則運算

Python 內建支援複數的加法、減法、乘法和除法，可以直接使用標準運算符 `+`, `-`, `*`, `/`。


```python
# 定義兩個複數
c1 = 2 + 3j
c2 = 1 - 2j

print(f"--- 複數四則運算範例 ---")
print(f"複數 c1 = {c1}")
print(f"複數 c2 = {c2}")

# 加法
sum_c = c1 + c2
print(f"\n加法: c1 + c2 = {sum_c}")

# 減法
diff_c = c1 - c2
print(f"減法: c1 - c2 = {diff_c}")

# 乘法
prod_c = c1 * c2
print(f"乘法: c1 * c2 = {prod_c}")

# 除法
div_c = c1 / c2
print(f"除法: c1 / c2 = {div_c}")
```

### 生成共軛複數

一個複數 `a + bj` 的共軛複數是 `a - bj`。Python 複數物件有一個 `.conjugate()` 方法來獲取其共軛複數。


```python
# 定義一個複數
c3 = 5 + 6j

print(f"--- 共軛複數範例 ---")
print(f"原始複數 c3 = {c3}")

# 獲取共軛複數
conj_c3 = c3.conjugate()
print(f"c3 的共軛複數 = {conj_c3}")

c4 = -2 - 7j
conj_c4 = c4.conjugate()
print(f"原始複數 c4 = {c4}")
print(f"c4 的共軛複數 = {conj_c4}")
```

### `cmath` 模組的進階複數運算

`cmath` 模組提供了許多用於複數的數學函數，這些函數在處理複數時非常有用，特別是在工程和科學計算中。

我們將示範以下幾個常用的進階函數：
*   `cmath.phase(z)`: 返回複數 `z` 的相位角（以弧度表示）。
*   `cmath.exp(z)`: 計算複數 `z` 的指數 $e^z$。
*   `cmath.log(z)`: 計算複數 `z` 的自然對數 $\ln(z)$。
*   `cmath.sqrt(z)`: 計算複數 `z` 的平方根。
*   `cmath.sin(z)`, `cmath.cos(z)`, `cmath.tan(z)`: 計算複數 `z` 的三角函數。


```python
import cmath
import math

print("--- cmath 進階複數運算範例 ---")

z_adv = 1 + 1j
print(f"處理的複數 z = {z_adv}")

# 1. 計算相位角 (Phase)
phase_z = cmath.phase(z_adv)
print(f"\n1. 相位角 (弧度): {phase_z:.4f} rad")
print(f"   相位角 (角度): {math.degrees(phase_z):.4f}°")

# 2. 指數運算 (Exponential)
# 根據歐拉公式 e^(ix) = cos(x) + i sin(x)
exp_z = cmath.exp(z_adv) # e^(1+1j) = e^1 * e^(1j) = e * (cos(1) + j*sin(1))
print(f"\n2. 指數 e^z: {exp_z:.4f}")

# 3. 自然對數 (Natural Logarithm)
# log(z) = log(|z|) + j*phase(z)
log_z = cmath.log(z_adv)
print(f"\n3. 自然對數 ln(z): {log_z:.4f}")

# 4. 平方根 (Square Root)
sqrt_z = cmath.sqrt(z_adv)
print(f"\n4. 平方根 sqrt(z): {sqrt_z:.4f}")
print(f"   驗證: {sqrt_z * sqrt_z:.4f}")

# 5. 三角函數 (Trigonometric Functions)
# cmath.sin(z)
sin_z = cmath.sin(z_adv)
print(f"\n5. 正弦 sin(z): {sin_z:.4f}")

# cmath.cos(z)
cos_z = cmath.cos(z_adv)
print(f"   餘弦 cos(z): {cos_z:.4f}")

# cmath.tan(z)
tan_z = cmath.tan(z_adv)
print(f"   正切 tan(z): {tan_z:.4f}")
print(f"   驗證 tan(z) = sin(z)/cos(z): {(sin_z / cos_z):.4f}")
```

課本範例：利用複數運算證明三角函數的和角公式 $\cos(\alpha+\beta)=\cos(\alpha)\cos(\beta)-\sin(\alpha)\sin(\beta)$  

## 複數指數訊號(Complex Exponential Signals)與相量(Phasor)
$z(t)=Ae^{jwt+\phi}$: 振幅$A$，相位角$wt+\phi$，相位移$\phi$  
相量：$Ae^{j\phi}$

### 相量加法
若 $\displaystyle \sum_{k=1}^N A_k \cos(w_0 t+\phi_k)= A\cos(w_0 t+ \phi)$.  
則 $\displaystyle \sum_{k=1}^N A_ke^{j\phi_k} = A e^{j\phi}$

### 相量加法合成弦波範例

假設我們有兩個頻率均為 10 Hz 的弦波：

*   **弦波 1 (y1):** 振幅 $A_1 = 5$，相位 $\phi_1 = 0$ 度
*   **弦波 2 (y2):** 振幅 $A_2 = 3$，相位 $\phi_2 = 90$ 度

我們將使用相量 (Phasor) 方法來合成這兩個弦波。

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

# Define parameters
f = 10      # Frequency in Hz
t = np.linspace(0, 1/f * 2, 500)  # Time array for two cycles

# --- Wave 1 parameters ---
A1 = 5
phi1_deg = 40
phi1_rad = np.deg2rad(phi1_deg)

# --- Wave 2 parameters ---
A2 = 3
phi2_deg = 120
phi2_rad = np.deg2rad(phi2_deg)

# 1. 將弦波轉換為相量 (Phasor)
# 我們將 A * sin(wt + phi) 的相量表示為 A * e^(j*phi)
Z1 = cmath.rect(A1, phi1_rad) # cmath.rect(r, phi) returns r * (cos(phi) + j*sin(phi))
Z2 = cmath.rect(A2, phi2_rad)

print(f"弦波 1 的相量 (Z1): {Z1:.4f}")
print(f"弦波 2 的相量 (Z2): {Z2:.4f}")

# 2. 相量加法
Z_sum = Z1 + Z2
print(f"合成後的相量 (Z_sum): {Z_sum:.4f}")

# 3. 將合成後的相量轉換回弦波形式
A_sum = abs(Z_sum)
phi_sum_rad = cmath.phase(Z_sum)
phi_sum_deg = np.rad2deg(phi_sum_rad)

print(f"\n合成弦波的振幅 (A_sum): {A_sum:.4f}")
print(f"合成弦波的相位 (phi_sum): {phi_sum_deg:.4f}° ({phi_sum_rad:.4f} 弧度)")

# 計算原始弦波和合成弦波的時域表示
y1 = A1 * np.sin(2 * np.pi * f * t + phi1_rad)
y2 = A2 * np.sin(2 * np.pi * f * t + phi2_rad)
y_sum = A_sum * np.sin(2 * np.pi * f * t + phi_sum_rad)

# 4. 繪製結果
plt.figure(figsize=(12, 7))
plt.plot(t, y1, label=f'Wave 1 (A={A1}, phi={phi1_deg}°)', linestyle='--')
plt.plot(t, y2, label=f'Wave 2 (A={A2}, phi={phi2_deg}°)', linestyle=':')
plt.plot(t, y_sum, label=f'Sum Wave (A={A_sum:.2f}, phi={phi_sum_deg:.2f}°)', linewidth=2, color='red')

plt.title('Sine Wave Synthesis using Phasor Addition')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.axhline(0, color='black', linewidth=0.5)
plt.show()
```
